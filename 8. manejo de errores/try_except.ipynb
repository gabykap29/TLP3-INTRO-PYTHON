# Manejo de Errores en Python

Cuando un programa encuentra una situación inesperada, Python lanza una **excepción**.
Si no la manejamos, el programa se detiene con un mensaje de error.

El manejo de errores nos permite **anticipar** esos problemas y responder de forma controlada.

---
## ¿Qué es una excepción?

Una excepción es una señal de que algo salió mal durante la ejecución del programa.

Ejemplos comunes:

| Excepción | Causa habitual |
|---|---|
| `ZeroDivisionError` | Dividir un número entre cero |
| `ValueError` | Pasar un valor de tipo correcto pero contenido inválido |
| `TypeError` | Operar con tipos incompatibles |
| `IndexError` | Acceder a un índice que no existe en una lista |
| `KeyError` | Acceder a una clave que no existe en un diccionario |
| `FileNotFoundError` | Intentar abrir un archivo que no existe |

---
## Sintaxis básica: `try` / `except`

```python
try:
    # código que puede fallar
except TipoDeError:
    # qué hacer si ese error ocurre
```

* El bloque `try` contiene el código **riesgoso**.
* El bloque `except` se ejecuta **solo si ocurre el error** indicado.
* Si no ocurre ningún error, el `except` se **omite**.

````python
try:
    div = 3 / 0
    print("try se ejecuto")
    
except ZeroDivisionError:
    print("Error, division por zero")
    
except TypeError: 
    print("Error de tipos")
except Exception as e:
    print("Todas las validaciones fallaron")
````

---
## Ejemplo 1 — `ZeroDivisionError`

```python
dividendo = 10
divisor = 0

try:
    resultado = dividendo / divisor
    print(resultado)
except ZeroDivisionError:
    print("Error: no se puede dividir por cero")
```

**Salida esperada:**
```
Error: no se puede dividir por cero
```

````python
try:
    div = 3 / 0
    print("try se ejecuto")
    
except ZeroDivisionError:
    print("Error, division por zero")
````

---
## Ejemplo 2 — `ValueError` al convertir input

Un caso muy frecuente: el usuario ingresa texto cuando se espera un número.

```python
entrada = "hola"

try:
    numero = int(entrada)
    print(f"El doble es: {numero * 2}")
except ValueError:
    print("Error: eso no es un número válido")
```

**Salida esperada:**
```
Error: eso no es un número válido
```

````python
try: 
    num = int(input("Ingrese un numero"))
    print(num)
except ValueError:
    print("Por favor, ingrese un numero entero!")
````

---
## Ejemplo 3 — Capturar múltiples excepciones

Se pueden tener varios bloques `except` para distintos tipos de error:

```python
lista = [10, 20, 30]
indice = "dos"

try:
    print(lista[indice])
except TypeError:
    print("Error: el índice debe ser un número entero")
except IndexError:
    print("Error: ese índice no existe en la lista")
```

**Salida esperada:**
```
Error: el índice debe ser un número entero
```

````python
try: 
    vector = [1,2,3,4]
    vector[6]
except IndexError:
    print("No existe ese index")
````

## El bloque `else`

El bloque `else` se ejecuta **solo si no ocurrió ningún error** en el `try`:

```python
try:
    # código riesgoso
except TipoDeError:
    # si hubo error
else:
    # si todo salió bien
```

```python
try:
    resultado = 10 / 2
except ZeroDivisionError:
    print("No se puede dividir por cero")
else:
    print(f"La división fue exitosa: {resultado}")
```

**Salida esperada:**
```
La división fue exitosa: 5.0
```

````python
try:
    print("SE ejecuto correctamente")
except Exception as e:
    print("Llego a la excepcion")
    
else:
    print("Se ejecuto el try correctamente")
````

---
## El bloque `finally`

El bloque `finally` se ejecuta **siempre**, haya error o no.
Es útil para liberar recursos (cerrar archivos, conexiones, etc.).

```python
try:
    numero = int("abc")
except ValueError:
    print("Error al convertir")
finally:
    print("Esto se ejecuta siempre")
```

**Salida esperada:**
```
Error al convertir
Esto se ejecuta siempre
```

````python
try:
    diccionario = {
        "nombre":"Gabriel",
        "edad": 27
    }
    diccionario["domicilio"]
except KeyError as e:
    print("Llego a la excepcion")
    
else:
    print("Se ejecuto el try correctamente")
finally:
    print("Se ejecuto todo el algoritmo")
````

---
## Estructura completa

```python
try:
    # código riesgoso
except TipoDeError:
    # si ocurrió ese error
else:
    # si no ocurrió ningún error
finally:
    # siempre se ejecuta
```

````python
try:
    diccionario = {
        "nombre":"Gabriel",
        "edad": 27
    }
    diccionario["domicilio"]
except KeyError as e:
    print("Llego a la excepcion")
    
else:
    print("Se ejecuto el try correctamente")
finally:
    print("Se ejecuto todo el algoritmo")
````

## Capturar el mensaje del error con `as`

Podemos obtener el mensaje de la excepción usando `as`:

```python
try:
    resultado = 10 / 0
except ZeroDivisionError as error:
    print(f"Ocurrió un error: {error}")
```

**Salida esperada:**
```
Ocurrió un error: division by zero
```

````python
try:
    print("Se ejecuto el try")
except Exception as e:
    print("Excepcion capturada")
    
````

````python

try:
    file = open("salud.txt")
    file.read()
except FileNotFoundError:
    print("Excepcion capturada")
````
